## Matplotlib

Low level graph plotting library that serves as a visualization utility.

Most of the utilities in matplotlib lies in `pyplot` submodule, which is usually imported under the alias `plt`.

The `plot()` function is used to draw points(markers) in a diagram.
By default, it draws a line from point to point. The function takes parameters for specifying the points.
- The parameter 1 is an array containing points in x axis
- The parameter 2 is an array containing points in y axis

```python
import matplotlib.pyplot as plt
import numpy as np

# Draw a line connecting (0, 0) and (8, 342)
x = np.array([0, 8])
y = np.array([0, 342])

plt.plot(x, y)
plt.show()
```

```python
# Plotting only the points ('o': 'rings')
plt.plot(x, y, 'o')
plt.show()
```

```python
# Connecting multiple points
x = np.array([1, 3, 5, 19, 29])
y = np.array([3, 5, 12, 3, 7])
plt.plot(x, y)
plt.show()
```

```python
# By default, the x points are taken as 0, 1, 2...

plt.plot(np.array([3, 5, 13, 16, 22, 23]))
plt.show()
```

### Markers

Use the keyword argument `marker` to emphasize each point with a specified marker. 

```python
ypoints = np.array([2, 0, 7, 12, 22])
plt.plot(ypoints, marker='*')
plt.show()
plt.plot(ypoints, marker='X')
plt.show()
```

### Format Strings `fmt`

Shortcut string notation to specify the marker.

< marker >< line >< color >

[Marker, Line and Color Reference](https://www.w3schools.com/python/matplotlib_markers.asp)

```python
# Triangle as marker, dotted line, red color
plt.plot(ypoints, '^:r')
plt.show()
```

```python
# Linestyle, ls -> dashed, '--'
ypts = np.array([3, 5, 6, 9, 2])
plt.plot(ypts, ls='--')
plt.show()
plt.plot(ypts, ls='--', color = 'red')
plt.show()
```

### Labels

Labelling Axes and Giving Title, also setting font properties for each.

```python
plt.plot(ypts)

font = {'family':'serif','color':'blue','size':20}

plt.title("Graph", fontdict=font)
plt.xlabel("Time")
plt.ylabel("Acceleration")

plt.show()
```

```python
# Adding grid lines
plt.plot(ypts)
plt.grid()
plt.show()

# Specifying axis of grid line
plt.plot(ypts)
plt.grid(axis='x')
plt.show()
```

### Subplot

Plotting multiple figures in one diagram.

The `subplot()` function takes 3 arguments
- First and Second argument represents the rows and columns the layout is organized in 
- Third argument represents the index of the current plot

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# Plot 1
plt.subplot(1, 2, 1)
plt.plot(x, y)

# Plot 2
plt.subplot(1, 2, 2)
plt.plot(y, x)

# Supertitle
plt.suptitle("Two in One")
plt.show()
```

## Other Types of Plots

```python
from numpy import random

x = random.randint(100, size=(100))
y = random.randint(100, size=(100))

# Scatter Diagram: Two plots in one diagram
plt.scatter(x, y)
plt.scatter(y, x)
plt.show()
```

```python
x = np.array(["a", "b", "c"])
y = np.array([34, 22, 156])

# Bar diagram
plt.bar(x, y)
plt.show()

# Horizontal bars
plt.barh(random.randint(10, size=(10)), random.randint(100, size=(10)), color="red")
plt.show()
```

```python
# Histograms
normal_dist = np.random.normal(170, 10, 250)
plt.hist(normal_dist)
plt.show()
```

```python
# Pie charts
portions = np.array([25, 50, 15, 5, 5])
labels = ["a", "b", "c", "d", "e"]
plt.pie(portions, labels=labels)

plt.show()
```

```python
# Specifying starting angle of the first wedge and also making "c" and "e" wedges stand out
plt.pie(portions, labels=labels, explode=[0, 0, 0.2, 0, 0.2])
plt.show()
```
